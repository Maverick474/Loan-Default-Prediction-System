```python
import pandas as pd
```

```python
df = pd.read_csv("original.csv")

# Checking for dataset if it's already cleaned, doesn't have any wrong data types and null values.
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
# Since the age has null values
df["age"] = df["age"].fillna(df["age"].median())

print(df.isnull().sum())
```

```python
# Dropping the client id because it is not needed 
df = df.drop("clientid", axis=1, errors="ignore")
print(df.head())
```

```python
print(df.shape)
print(df.duplicated().sum())
print(df["default"].value_counts())
print(df.describe())
```

```python
# Since there was an unusual value in age column (Outlier) we need to fix it
df = df[df["age"] > 0]
```

```python
print(df.describe())
```

```python
print(df.duplicated().sum())

print(df["default"].value_counts())
```

```python
# Saving the clean dataset

df.to_csv("cleaned_loan_dataset.csv", index=False)
```
